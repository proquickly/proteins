# How to
I read the pickle files and load them in to pandas dataframes for further analysis

1. Run load (menu option 1, then 2).
1. Then run this. The notebook uses the pickle files that load.py creates. Therefore you only need to run load, which takes a long time, if you want to refresh the data.
